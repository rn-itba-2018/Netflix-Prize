# Netflix Prize

El premio Netflix fue un desafío que consistió en mejorar el sistema de recomendaciones de películas disponibles. La empresa proveyó una base de datos de 100 millones de puntuaciones que 480 mil usuarios le dieron a 18 mil películas. El objetivo es entrenar un predictor y minimizar el error cuadrático para 1,4 millones de puntuaciones que los participantes desconocían.

A partir de un dataset reducido vamos a replicar el desafío. El archivo 'data_set.csv' contiene 80 mil puntuaciones de mil usuarios para mil setecientas películas, con el siguiente formato:

id_usuario,id_película,puntaje,unix_timestamp

También se otorgará un archivo .npy con un array numpy ya procesado de los datos donde la fila indicará el usuario (usuario 1 en el csv equivale a fila con índice 0). La columna indicará la película y el numero que figura en dicha matríz indicará el puntaje entre 1 y 5. De figurar un 0 significará que no se tiene el dato.

El alumno puede utilizar el dato que más le convenga para realizar sus cálculos.

## Entrega de Datos

Se adjuntará también un archivo llamado 'a_puntuar.csv' que tiene 20 mil puntuaciones a predecir, con el siguiente formato:
    
id_usuario,id_película,unix_timestamp

El alumno deberá generar un archivo llamado 'answer.csv' con la misma estructura que 'data_set.csv' que contenga la misma información que 'a_puntuar.csv' pero con una columna extra en el que figure su puntaje.

Se deberá entregar:

- Archivo 'answer.csv'
- Todo el código que el alumno haya utilizado para generar dicho archivo. Incluyendo el entrenamiento de la red y el código en la cual utiliza la red para predecir

### Nota del Trabajo

Si bien el resultado final obtenido (mse como explicado en la sección 1.2.2) es importante, no es lo único que se tendrá en cuenta en la nota.
Para evaluar se tendrán en cuenta diversos factores. Entre ellos (y principalmente) se observará el código y se verificará cómo hizo el alumno para encarar el problema. 

Funciones extra que no se utilizen o carezcan de sentido restará puntos. 
La prolijidad del código también afectará la nota, si bien de forma sutil.

Será de valor que, si el alumno utiliza algún paper o aplica alguna técnica que valga la pena mencionar, lo comente de alguna forma en un archivo txt o en una notebook jupyter. Es importante ya que de aplicarse algún algoritmo poco convencional que podría sumar mucho puntaje, de no estar explicado no será tenido en cuenta (más que su reflejo en el resultado).

También será tenido en cuenta si el alumno decide realizar una implementación que no funcionó bien o que luego terminó no usandola para el código final ya que demuestra una investigación más profunda por parte del alumno. En dicho caso también deberá figurar una explicación lo que se hizo y qué otras estrategias se probaron.

### Desafío

Se realizará el MSE de los valores obtenidos por el alumno y los valores reales de los datos. El alumno que consiga una mejor MSE obtendrá los puntos extra del desafío.
